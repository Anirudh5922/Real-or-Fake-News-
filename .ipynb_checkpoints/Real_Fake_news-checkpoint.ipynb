```python
def clean_text(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub('https?://\S+|www\.\S+','',text)
    text=re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    text = re.sub('\n', ' ', text)
    text = re.sub('\w*\d\w*', '', text)
    tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
    text=tokenizer.tokenize(text)
    return ' '.join(text)
```

```python
%%time
def remove_stopwords(text):
    new_text=[]
    for i in text.split():
        if i.strip().lower() not in stop and len(i.strip())>1:
            new_text.append(i.strip())
    return " ".join(new_text)
text=text.apply(remove_stopwords)
```

```python
x_train,x_test,y_train,y_test = train_test_split(data.text,data.target,random_state = 21)
```

```python
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
max_features=10000
maxlen=900
```

```python
from tensorflow.keras.preprocessing import text
```

```python
tokenizer=text.Tokenizer(num_words=max_features)
tokenizer.fit_on_texts(x_train)
tokenized_train=tokenizer.texts_to_sequences(x_train)
x_train=sequence.pad_sequences(tokenized_train,maxlen=maxlen) 
```

```python
x_train
```

```python
tokenized_test=tokenizer.texts_to_sequences(x_test)
x_test=sequence.pad_sequences(tokenized_test,maxlen=maxlen)
```

```python
x_test
```

```python
EMBEDDING_FILE = '../input/glovetwitter27b100dtxt/glove.twitter.27B.100d.txt'
```

```python
def get_coefs(word, *arr): 
    return word, np.asarray(arr, dtype='float32')
embeddings_index = dict(get_coefs(*o.rstrip().rsplit(' ')) for o in open(EMBEDDING_FILE,encoding="utf8"))
```

```python
all_embs = np.stack(embeddings_index.values())
emb_mean,emb_std = all_embs.mean(), all_embs.std()
embed_size = all_embs.shape[1]

word_index=tokenizer.word_index
n_words=min(max_features,len(word_index))
embedding_matrix = np.random.normal(emb_mean, emb_std, (n_words, embed_size))
for word, i in word_index.items():
    if i >= max_features: continue
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None: embedding_matrix[i] = embedding_vector
```

```python
embedding_matrix
```

```python
batch_size =256
epochs = 10
embed_size =100
```

```python
lstm_model = Sequential()
lstm_model.add(Embedding(max_features,output_dim=embed_size,weights
         =[embedding_matrix],input_length=maxlen,trainable=False))
lstm_model.add(LSTM(units=128 , return_sequences = True))
lstm_model.add(Dropout(0.2))
lstm_model.add(LSTM(units=64))
lstm_model.add(Dropout(0.1))
lstm_model.add(Dense(units=32,activation='relu'))
lstm_model.add(Dense(1,activation='sigmoid'))
lstm_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
lstm_model.summary()
```

```python
lstm_model.fit(x_train,y_train,batch_size=batch_size,
               validation_data=(x_test,y_test),epochs=epochs)
```

```python
lstm_model.save("lstm.h5")
```
